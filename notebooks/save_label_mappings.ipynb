```python
# importing the required libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
import os
import joblib
from sklearn.preprocessing import MinMaxScaler
```

```python
combined_df = pd.read_csv("../data/customer_features.csv")
```

```python
combined_df.head()
```

```python
# Dictionary to store the mapping (original value: encoded number) for each column
mappings = {}

name_encoder = LabelEncoder()
for col in ['gender', 'paymentmethod', 'industry']:
    # Fit and transform the current column
    combined_df[col] = name_encoder.fit_transform(combined_df[col])

    # Store the mapping *before* the encoder is overwritten in the next iteration
    # Create a dictionary mapping the original class names (classes_) to their index (0, 1, 2, ...)
    class_names = name_encoder.classes_
    mapping = {class_names[i]: i for i in range(len(class_names))}
    mappings[col] = mapping
    
    print(f"\n--- Mapping for '{col}' ---")
    print(mapping)

# Access the mapping later
print("\n--- Final Mappings Dictionary ---")
print(mappings['gender'])
# Example: Check which value is 0 for 'gender'
# Note: You can reverse the dictionary if you want to check 0 = ?
```

```python
save_path = "../models"
os.makedirs(save_path, exist_ok=True)
```

```python
# saving name_encoder

try:
    joblib.dump(name_encoder, os.path.join(save_path, "name_encoder.pkl"))
    print("Encoder saved")
except Exception as e:
    print("Encoder not found or not saved")
```

```python
# saving the mappings 
joblib.dump(mappings, "../models/mappings.pkl")
```
